```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def get_cosine_similarity(sentence1, sentence2):
    # Tokenize and convert the sentences into numerical vectors
    vectorizer = CountVectorizer().fit_transform([sentence1, sentence2])

    # Calculate the cosine similarity between the vectors
    cosine_sim = cosine_similarity(vectorizer)

    # The cosine similarity is in the upper-right corner of the matrix
    return cosine_sim[0, 1]

# Example usage:
clause1 = "Pejabat Pengelola BLU terdiri atas Pemimpin, Pejabat Keuangan; dan Pejabat Teknis"
clause2 = "Pemimpin BLU melakukan pengawasan dan pengendalian terhadap pengelolaan aset pada BLU yang berada dalam penguasaannya"

similarity = get_cosine_similarity(clause1, clause2)
print(f"Cosine Similarity: {similarity}")
```
